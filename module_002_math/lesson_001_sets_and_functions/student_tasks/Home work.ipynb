# Задание 1:
Выполнить операции: $A \cup B; A \cap B; A \setminus B; B \setminus A; A \times B; B \times A$, где A и B

* $A=\{a,1,2\};B=\{a,b,1\}$
* $A=\{2n-1 \mid n \in N\}; B=\{-1,0,1,2,3\}$
* $A=(-\infty , 3); B=[-1, +\infty)$

```python
A = {'a', 1, 2}
B = {'a', 'b', 1}

print('A:', A)
print('B:', B)

print('Union', A.union(B))
print('Intersection', A.intersection(B))
print('Difference_AB', A.difference(B))
print('Difference_BA', B.difference(A))

from itertools import product
A_B = [str(A)+str(B) for A, B in product(A, B)]
B_A = [str(B)+str(A) for A, B in product(B, A)]

print('AxB', A_B)
print('BxA', B_A)
print('\n--------------\n')


X = {1}
n = 100
for i in range(1, n-1):
    X.add(2*i - 1)

print('X:', X)
Y = {-1, 0, 1, 2, 3}
print('Y:', Y)

X_Y = [str(X)+str(Y) for X, Y in product(X, Y)]
Y_X = [str(Y)+str(X) for X, Y in product(Y, X)]

print('Union', X.union(Y))
print('Intersection', X.intersection(Y))
print('Difference_AB', X.difference(Y))
print('Difference_BA', Y.difference(X))
print('XxY', X_Y)
print('YxX', Y_X)
print('\n--------------\n')



```

# Задание 2:
Построить графики всех элементарных функций с данного сайта: http://mathprofi.ru/grafiki_i_svoistva_funkcij.html

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np


x = np.arange(-10, 10, 5)
print('x:', x)
y = np.arange(-10, 10, 5)
print('y:', y)

plt.plot(x, y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Line')
plt.grid(True)
plt.show()


def parabola(x):
    return x*x

x = np.arange(-10, 10, 0.5)
print('x:', x)
y = parabola(x)
print('y:', y)

plt.plot(x, y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Parabola')
plt.grid(True)
plt.show()


def CUBparabola(x):
    return x**3

x = np.arange(-3, 3, 0.2)
print('x:', x)
y = CUBparabola(x)
print('y:', y)

plt.plot(x, y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('CUBparabola')
plt.grid(True)
plt.show()


def Sqrt_x(x):
    return np.sqrt(x)

x = np.arange(0, 10, 0.2)
print('x:', x)
y = Sqrt_x(x)
print('y:', y)

plt.plot(x, y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Sqrt_x')
plt.grid(True)
plt.show()


def Giper(x):
    return 1/x

x = np.arange(-5, 5, 0.3)
print('x:', x)
y = Giper(x)
print('y:', y)

plt.plot(x, y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Giper')
plt.grid(True)
plt.show()


def Exp_x(x):
    return np.exp(x)

x = np.arange(-10, 10, 0.2)
print('x:', x)
y = Exp_x(x)
print('y:', y)

plt.plot(x, y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Exp_x')
plt.grid(True)
plt.show()


def Ln_x(x):
    return np.log(x)

x = np.arange(0.1, 10, 0.1)
print('x:', x)
y = Ln_x(x)
print('y:', y)

plt.plot(x, y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Ln_x')
plt.grid(True)
plt.show()


def Sin(x):
    return np.sin(x)

x = np.arange(-10, 10, 0.2)
print('x:', x)
y = Sin(x)
print('y:', y)

plt.plot(x, y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Sin')
plt.grid(True)
plt.show()


def Cos(x):
    return np.cos(x)

x = np.arange(-10, 10, 0.2)
print('x:', x)
y = Cos(x)
print('y:', y)

plt.plot(x, y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Cos')
plt.grid(True)
plt.show()


def Tan(x):
    return np.tan(x)

x = np.arange(-4, 4, 0.05)
print('x:', x)
y = Tan(x)
print('y:', y)

plt.plot(x, y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Tan')
plt.grid(True)
plt.show()


def Arcsin(x):
    return np.arcsin(x)

x = np.arange(-1, 1, 0.02)
print('x:', x)
y = Arcsin(x)
print('y:', y)

plt.plot(x, y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Arcsin')
plt.grid(True)
plt.show()


def Arccos(x):
    return np.arccos(x)

x = np.arange(-1, 1, 0.02)
print('x:', x)
y = Arccos(x)
print('y:', y)

plt.plot(x, y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Arccos')
plt.grid(True)
plt.show()


def Arctan(x):
    return np.arctan(x)

x = np.arange(-6, 6, 0.05)
print('x:', x)
y = Arctan(x)
print('y:', y)

plt.plot(x, y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Arctan')
plt.grid(True)
plt.show()



```
