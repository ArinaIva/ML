```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

def showPlot(x_start, x_max, step, callback, title='Function'):
    x = np.arange(x_start, x_max, step)
    y = callback(x)

    plt.plot(x, y)
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.title(title)
    plt.grid(True)
    plt.show()


from sympy import *

x, y = symbols('x y')

def value_of_expr(x_vector, symbol, expr):
    y = np.empty_like(x_vector)
    for index, element  in enumerate(x_vector):
        y[index] = (expr.subs(symbol, element))
    return y


e = x**3+3*x**2-72*x+1

showPlot(-10, 10, 0.1, lambda i: value_of_expr(i, x, e), title='Function')

dx = diff(e, x)
dx

showPlot(-10, 10, 0.1, lambda i: value_of_expr(i, x, dx), title='Diff1')
```

```python
h = log(2*x-1)

showPlot(0.6, 10, 0.1, lambda i: value_of_expr(i, x, h), title='Function')

dx = diff(h, x)
dx

showPlot(0.6, 10, 0.1, lambda i: value_of_expr(i, x, dx), title='Diff1')

second_dex = diff(h, x, x)
second_dex

showPlot(0.6, 10, 0.1, lambda i: value_of_expr(i, x, second_dex), title='Diff2')

third_dex = diff(h, x, x, x)
third_dex

showPlot(0.6, 10, 0.1, lambda i: value_of_expr(i, x, third_dex), title='Diff3')

forth_dex = diff(h, x, x, x, x)
forth_dex

showPlot(0.6, 10, 0.1, lambda i: value_of_expr(i, x, forth_dex), title='Diff4')

tenth_dex = diff(h, x, x, x, x, x, x, x, x, x, x)
tenth_dex

showPlot(0.6, 10, 0.1, lambda i: value_of_expr(i, x, tenth_dex), title='Diff10')



```
