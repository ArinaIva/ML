# Задача 2
Анализ данных. Dataset - https://www.kaggle.com/ronitf/heart-disease-uci/version/1. Из приложенного файла(heart.csv) загружается dataset о наличии сердечных заболеваний.
Найдите следующие ответы:
* Какой средний возраст женщин у которых присутсвуют сердечные заболевания
* Какой самый распространненый вид боли в груди при наличии сердечных заболеваний
* Подумать. Попробуйте определить влияет ли ЭКГ(restecg) на наличие сердечных заболеваний, сделайте выводы


```python
import csv
dataset = []
with open('../datasets/heart.csv', newline='') as csvfile:
    spamreader = csv.reader(csvfile, delimiter=',', quotechar='|')
    next(spamreader, None) # skip header
    for row in spamreader:
        dataset.append(list(map(lambda x: int(x) if x.find('.') == -1 else float(x), row)))
```

```python
print('количество пациентов', len(dataset), '\n_____________________________\n')

for index, value in enumerate(dataset, start=1):
    print(index,'__', value)
    
print('\n_____________________________\n')
    
female_ages = []
for i in dataset:
    if i[1] == 0 and i[13] == 1:
        female_ages.append(i[0])  

print('Средний возраст женщин обладающих середечными заболеваниями: ',sum(female_ages)/len(female_ages))

print('\n_____________________________\n')

chest_pain_0 = []
chest_pain_1 = []
chest_pain_2 = []
chest_pain_3 = []

for i in dataset:
    if i[13] == 1:
        if i[2] == 0:
            chest_pain_0.append(i[2])
        elif i[2] == 1:
            chest_pain_1.append(i[2])
        elif i[2] == 2:
            chest_pain_2.append(i[2])
        elif i[2] == 3:
            chest_pain_3.append(i[2])
        
print(len(chest_pain_0))
print(len(chest_pain_1))
print(len(chest_pain_2))
print(len(chest_pain_3))


a = [chest_pain_0, chest_pain_1, chest_pain_2, chest_pain_3]
b = a[0]
for i in range(1,4):
    if len(b) < len(a[i]):
        b=a[i]

print('самый распространненый вид боли в груди при наличии сердечных заболеваний', b[0])
print('\n_____________________________\n')


print('Влияют ли показания ЭКГ(restecg) на наличие сердечных заболеваний?')

true = []
false = []
for i in dataset:
    if i[13] == 0 and i[6] == 0:
        true.append('+')
    elif i[13] == 1 and i[6] == 1 or i[6] == 2:
        true.append('+')
    else:
        false.append('-')
        
if len(true) > len(false):
    print('Влияют, ЭКГ - топчик')
else:
    print('Не влияют, ЭКГ - днище')
```
